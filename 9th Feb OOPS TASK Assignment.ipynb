```python
#OOPS TASK 
"""Q1. Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle."""
Ans: class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle
        
#Example car = Vehicle("Honda Civic", 120, 30)
        
"""Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity."""
Ans: class Car(Vehicle):
    def __init__(self, name, color, model, year):
        super().__init__(name, color, model, year)

    def seating_capacity(self, capacity):
        return f"{self.name} has a seating capacity of {capacity}."
    
    # Create an instance of Car
car = Car("Honda Civic", "Silver", "2020", "2020")

# Print the seating capacity of the car
print(car.seating_capacity(5)) # Honda Civic has a seating capacity of 5.

"""Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance."""
Ans: Multiple inheritance is a feature of object-oriented programming languages in which a class can inherit characteristics and features from more than one parent class.

Example:

class Parent1: 
    def __init__(self): 
        self.str1 = "Parent1"
        print("Parent1 constructor")
  
class Parent2: 
    def __init__(self): 
        self.str2 = "Parent2"        
        print("Parent2 constructor")
  
class Child(Parent1, Parent2): 
    def __init__(self): 
          
        # Calling constructors of Parent1 
        # and Parent2 classes 
        Parent1.__init__(self) 
        Parent2.__init__(self) 
        print("Child constructor")
  
    def printStrs(self): 
        print(self.str1, self.str2) 
  
ob = Child() 
ob.printStrs()

#Output:
Parent1 constructor
Parent2 constructor
Child constructor
Parent1 Parent2

"""Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class."""Getters and setters are methods used to access and modify the values of an object's properties. They are used to ensure that the data stored in an object is valid and secure.

In Python, getters and setters are created using the @property decorator.


class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        if not isinstance(value, str):
            raise TypeError('Please provide a string value for the name')
        self._name = value

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if not isinstance(value, int):
            raise TypeError('Please provide an integer value for the age')
        self._age = value

person = Person('John', 25)
print(person.name) # John
print(person.age) # 25
person.name = 'Jane'
person.age = 30
print(person.name) # Jane
print(person.age) # 30
person.name = 123 # TypeError: Please provide a string value for the name
person.age = '30' # TypeError: Please provide an integer value for the age

"""Q5.What is method overriding in python? Write a python code to demonstrate method overriding."""
Ans: Method overriding is a feature of object-oriented programming that allows a subclass or child class to provide a specific 
implementation of a method that is already provided by one of its superclasses or parent classes. It is used to override or 
replace the functionality of an existing method.

In Python, method overriding occurs when subclasses define methods with the same name as one of their superclasses.

Example:

class Parent: 
    def greet(self): 
        print("Hello") 
  
class Child(Parent): 
    def greet(self): 
        print("Hi") 
  
obj = Child() 
obj.greet() 

# Output: Hi
```
